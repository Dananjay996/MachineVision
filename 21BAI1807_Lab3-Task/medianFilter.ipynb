```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = plt.imread("./image.png")
plt.imshow(image)
```

```python
#Apply median filter to the image without using any external libraries
def median_filter(image, kernel_size):
    #Get the dimensions of the image
    height, width, channels = image.shape
    #Create a new image to store the filtered image
    new_image = np.zeros((height, width, channels))
    #Get the padding size
    padding = kernel_size // 2
    #Iterate through the image
    for i in range(padding, height - padding):
        for j in range(padding, width - padding):
            #Get the kernel
            kernel = image[i - padding:i + padding + 1, j - padding:j + padding + 1]
            #Get the median of the kernel
            new_image[i, j] = np.median(kernel, axis=(0, 1))
    return new_image
```

```python
filtered_image = median_filter(image,3)
plt.imshow(filtered_image)
```
