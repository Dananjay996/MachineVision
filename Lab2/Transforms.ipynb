```python
import matplotlib.pyplot as plt
import numpy as np

image = plt.imread("../images/colorImage.webp")
plt.imshow(image, interpolation=None)
print(image)
```

```python
def perform_transformation(img):
    inverted_img = 255 - img
    return inverted_img
```

```python
img = perform_transformation(image)
plt.imshow(img, interpolation=None)
```

```python
#Logarithmic transformation
def perform_log_transformation(img,c):
    log_transformed_img = c * np.log(1 + img)
    log_transformed_img = np.clip(log_transformed_img, 0, 255)
    log_transformed_img = log_transformed_img.astype(np.uint8)
    return log_transformed_img
```

```python
# image = image.copy()
# image = image.astype(np.float32)
log_image = perform_log_transformation(image,50)
plt.imshow(log_image, interpolation=None)
```

```python
# Gamma transformation
def perform_gamma_transformation(img, gamma):
    gamma_transformed_img = np.power(img, gamma)
    gamma_transformed_img = np.clip(gamma_transformed_img, 0, 255)
    gamma_transformed_img = gamma_transformed_img.astype(np.uint8)
    return gamma_transformed_img
```

```python
gamma_img = perform_gamma_transformation(image, 2)
plt.imshow(gamma_img, interpolation=None)
```
